## 問2.5
draw 関数を用いて x = -10 から x = 10 までの x に対するベクトル(x, x**2) を点で描け

```python
draw(
    Points(*[(x, x**2) for x in range(-10, 11)]),
    grid=(1, 10),
    nice_aspect_ratio=False
)
```
